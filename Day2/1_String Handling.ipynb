```python
#Strings are one of the commonly used datatype ideally used to store text data
```

```python
#1. Creating Strings
#
# Single Quote

string1 = 'Hello World'
print(string1)

# Double Quote
string2 = "Hello world. This is Prashant's class"
print(string2)

#Triple Single Quote
string3 = '''
Welcome to Python
We are learning String Handling
'''
print(string3)
```

```python
#Accessing elements in string

#Access single character
print(string2[6])

#Accessing a substring
print(string2[6:11])
```

```python
#Basic String Operations

# Concatenation -------- Merging of two or more strings

var1 = "Hello"
var2 = "World"

print(var1 + var2)

print(var1 + " " + var2)


#Repetition (*)

print(var1 * 3)

temp = var1+" "
print(temp * 3)


#Search Operation

data = "Welcome to the world of Python. You are learning String Handling"

#Check if the word Python exists in the data variable
print("Python" in data)
```

```python
#String Methods

# Changing Case (Normalization)

print(data.upper())
print(data.lower())
```
