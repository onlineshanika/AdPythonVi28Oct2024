```python
# Conditional Statements in Python
#
# 1. if statement
# 2. if..else statement
# 3. elif ladder
```

```python
# if statement
#
# In this expression we can supply only ONE condition. If the condition is met, it will execute the code block else ignore code block and execute
# further code instructions
#
# if expression/condition:
#     statement/block
#
#
# if condition {
#
# }
#


num = int(input("Enter a number: "))

if num > 10 :
  print("Greater than 10")
  print("Greater than 10")

print("Code Ended")
```

```python
#if .. else statement
#
# In this expression we can supply binary condition (two conditions)
#
# Syntax:
#
#  if expression1:
#        action1
#  else:
#        action2

num = int(input("Enter a number: "))

if num > 10 :
  print("Greater than 10")
else:
  print("Less than 10")

print("Code Ended")
```

```python
# elif ladder
#
# In this expression, you can supply multiple condition

num = int(input("Enter a number: "))

if num == 1:
  print("One")
elif num == 2:
  print("Two")
elif num == 3:
  print("Three")
elif num == 4:
  print("Four")
elif num == 5:
  print("Five")
else:
  print("I dont know")
```

```python
num = int(input("Enter a number: "))
if num == 1 or num == 2:
  print("Yes")
else:
  print("No")
```
