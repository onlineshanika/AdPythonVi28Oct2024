```python
import sqlite3
```

```python
#Establish Connection with DB

conn = sqlite3.connect('test.db')
```

```python
#Create Table

c = conn.cursor()
c.execute('''

CREATE TABLE employee (eid int, ename text, esal int, dept text)

''')
```

```python
#Insert Data

c.execute('''

INSERT INTO employee values (1,'Prashant',2112121,'Ops')

''')
```

```python
c = conn.cursor()
c.execute('''
select * from employee
''')
```

```python
#Commit the operation
conn.commit()
conn.close()
```
