```python
# File Handling
# 1. Read a file
# 2. Write data into a file
# 3. Using with
```

```python
#Built-in library to work with files ---------------------- open()
#
# Syntax:
#  open(fileLocation, mode)
#
# Where mode supported by open() is
#   r ------------- read
#   w ------------- write
#   a ------------- append
#   r+ ------------ reading and writing
```

```python
#opening a file

f = open('state_of_union.txt','r')

# Close the file

f.close()
```

```python
#opening a file

f = open('state_of_union.txt','r')

#To view content

print(f.read())

# Close the file

f.close()
```

```python
#with block: Using this you can access the file for any kind of operation
#             Once the operation is done, the file gets automatically closed.

with open('state_of_union.txt','r') as f:
  print(f.read())
```

```python
#Writing file

with open("test.txt",'w') as file1:
  file1.write("Hello World")
```

```python

#For http/https -- requests package
# For ftp -------- ftplib
# ssh key based logins ----paramiko

import requests

url = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"

response = requests.get(url)

if response.status_code == 200:
  content = response.text
  print(content)
else:
  print("Error reading file")
```

```python
import paramiko

hostname = 'sftp.example.com'
username = 'username'
password = 'password'

# Create an SFTP session
client = paramiko.SSHClient()
client.set_missing_host_key_policy(paramiko.AutoAddPolicy())
client.connect(hostname, username=username, password=password)

# Open an SFTP session
sftp = client.open_sftp()
remote_file = sftp.open('/path/to/remote_file.txt')
content = remote_file.read().decode('utf-8')  # Read and decode text

print(content)

# Close connections
remote_file.close()
sftp.close()
client.close()
```
