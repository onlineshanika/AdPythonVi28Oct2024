```python
# 1. Try-except block
#
# Syntax:
#  try:
#     statement which needs to be checked
#  except:
#     error to be given by user and terminate program normally
```

```python
try:
  10 / 0
except:
  print("You cannot divide by zero")
```

```python
#Addition of two numbers
try:
  num1 = int(input("Enter num1: "))
  num2 = int(input("Enter num2: "))
  print(f"Addition of {num1} and {num2} is {num1+num2}")
except:
  print("Enter numbers only")


```

```python
try:
  d = {"Maharashtra":"Mumbai"}
  print(d['Maharashtra'])
  d[[1,2,3]]="value"
except KeyError:
  print("Key Doesnt exists")
except TypeError:
  print("Dont use unhashable datatype. Instead use primitive datatypes")
```

```python
#The else clause ---- execute the code block if no exception occured
#
#

#Addition of two numbers
try:
  num1 = int(input("Enter num1: "))
  num2 = int(input("Enter num2: "))
except:
  print("Enter numbers only")
else:
  print(f"Addition of {num1} and {num2} is {num1+num2}")
```

```python
# 3. the finally clause ---- Execute the code regardless of excpetions
#Addition of two numbers
try:
  num1 = int(input("Enter num1: "))
  num2 = int(input("Enter num2: "))
except:
  print("Enter numbers only")
else:
  print(f"Addition of {num1} and {num2} is {num1+num2}")
finally:
  print("Program Terminated")
```
